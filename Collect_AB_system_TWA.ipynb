# Model

\begin{align*}
H & =\Delta_{{\rm M}}b^{\dagger}b+\Omega\left(S_{x}^{A}-S_{x}^{B}\right)+\delta\left(S_{z}^{A}-S_{z}^{B}\right)+g\left(\left(S_{A}^{+}+S_{B}^{+}\right)b+\text{h.c.}\right).
\end{align*}

# Dynamics

```python
import time

t0=time.time()
#Parameters
import numpy as np
from numpy import sqrt

##Number of atoms

N_atom=100

##Basic parameters


gC=1
DeltaM=20*gC*np.sqrt(N_atom/2)
kappa=0  #keep this zero to model the unitary evolution

Omega_c = (N_atom*  gC**2)/sqrt(DeltaM**2+(kappa**2)/4)   ## I used N---N/2 because the effective number per side is N/2

delta=3*Omega_c


OmegaX=0.0*Omega_c
```
